```python
# make me 4 columns as HDD, SSD, Hybrid and Flash_Storage if df["memory" ] have SSD write that integer in SSD (for example 128 SSD in SSD column give value as 128 and 0 in other columns)
# convert GB="" and TB=1000  for example 2TB HDD', '64GB SSD 2000 in HDD and 64 in SSD
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame
# Make a copy to avoid SettingWithCopyWarning
df_processed = df.copy()

# 1. Clean and standardize the 'Memory' column
# Remove '.0' from values like '1.0TB'
df_processed['Memory'] = df_processed['Memory'].astype(str).replace('\.0', '', regex=True)
# Replace 'GB' with an empty string
df_processed['Memory'] = df_processed['Memory'].str.replace('GB', '')
# Replace 'TB' with '000' to convert to GB (1TB = 1000GB)
df_processed['Memory'] = df_processed['Memory'].str.replace('TB', '000')
# Rename 'Flash Storage' to match the desired column name
df_processed['Memory'] = df_processed['Memory'].str.replace('Flash Storage', 'Flash_Storage')

# 2. Initialize the four new columns with a default value of 0
df_processed['HDD'] = 0
df_processed['SSD'] = 0
df_processed['Hybrid'] = 0
df_processed['Flash_Storage'] = 0

# 3. Process each string in the 'Memory' column
# Split the string by ' + ' in case of dual drives
for i, memory_string in enumerate(df_processed['Memory']):
    parts = memory_string.split(' + ')
    for part in parts:
        # Split the part into value and storage type (e.g., '128 SSD')
        components = part.strip().split()
        if len(components) == 2:
            value, storage_type = components
            # Find the correct column and add the value
            if storage_type in ['HDD', 'SSD', 'Hybrid', 'Flash_Storage']:
                # Use .loc to safely assign the value to the correct row and column
                df_processed.loc[i, storage_type] += int(value)

df=df_processed.copy()

# You can now drop the original 'Memory' column if you wish
# df_processed.drop(columns=['Memory'], inplace=True)
```

## Model Training and Model Selection

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
## Create function to evaluate model
def evaluate_model(true,predict):
    mae=mean_absolute_error(true,predict)
    mse=mean_squared_error(true,predict)
    rmse=root_mean_squared_error(true,predict)
    r2_sqaure=r2_score(true,predict)
    return mae,mse,rmse,r2_sqaure
```

```python
models = {
                "Linear Regression": LinearRegression(),
                "Ridge":Ridge(),
                "Lasso":Lasso(),
                "K-Neighbors Regressor": KNeighborsRegressor(),
                "Random Forest": RandomForestRegressor(),
                "Decision Tree": DecisionTreeRegressor(),
                "Gradient Boosting": GradientBoostingRegressor(),
                "AdaBoost Regressor": AdaBoostRegressor(),
                "SVR":SVR(),
                "XGBRegressor":XGBRegressor()
        }
```

```python
try:
        from sklearn.pipeline import Pipeline
        preprocessor=ColumnTransformer(transformers=[
        ("cl_tnf",OneHotEncoder(sparse_output=False,drop="first"),[0,1,7,10,11])
        ],remainder="passthrough")

        X_transformed=preprocessor.fit_transform(X)
        from sklearn.model_selection import train_test_split
        X_train,X_test,y_train,y_test=train_test_split(X_transformed,y,test_size=0.15,random_state=2)


        r2_list=[]
        model_list=[]
        for i in range(len(list(models))):
                model=list(models.values())[i]
                model.fit(X_train,y_train)

         #Make predicitons
                y_train_pred=model.predict(X_train)
                y_test_pred=model.predict(X_test)

        # Evaluate Train and Test dataset
                model_train_mae,model_train_mse, model_train_rmse, model_train_r2 = evaluate_model(y_train,y_train_pred)
                model_test_mae,model_test_mse, model_test_rmse, model_test_r2 = evaluate_model(y_test,y_test_pred)
        
                print(list(models.keys())[i])
                model_list.append(list(models.keys())[i])

                print("Model preformance for Training set")
                print("Root Mean Squared Error: ",model_train_rmse)
                print("Mean Absolute Error: ",model_train_mae)
                print("R2 Score: ",model_train_r2)

                print("--------------------------")

                print("Model preformance for Test set")
                print("Root Mean Squared Error: ",model_test_rmse)
                print("Mean Absolute Error: ",model_test_mae)
                print("R2 Score: ",model_test_r2)
                r2_list.append(model_test_r2)

                print('='*35)
                print('\n')
                
except Exception as e:
        raise e
```
